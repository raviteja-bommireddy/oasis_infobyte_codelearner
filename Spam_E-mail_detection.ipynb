```python
df=pd.read_excel("Spam Email Detection.xlsx")
```

```python
print(df)
```

```python
df.drop(df.columns[2:6], axis=1, inplace=True)
print(df)
```

```python
df.info()
```

```python
df.shape
```

```python
df.loc[df['v1']=='spam','v1',]=0
df.loc[df['v1']=='ham','v1',]=1
```

```python
category=df['v1']
message=df['v2']
```

```python
print(category)
```

```python
print(message)
```

```python
category_train,category_test,message_train,message_test=train_test_split(category,message,test_size=0.2,random_state=3)
#from this line divided dataset into test,train dataset with 80% training , 20% testing
```

```python
print(category.shape)
print(category_train.shape)
print(category_test.shape)
print(message.shape)
print(message_train.shape)
print(message_test.shape)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
# Convert integer data to strings
message_train = [str(msg) for msg in message_train]
# Initialize the TfidfVectorizer with appropriate parameters
feature_extraction = TfidfVectorizer(min_df=1, stop_words='english', lowercase=True)

# Transform the training messages into feature vectors
message_train_features = feature_extraction.fit_transform(message_train)

# Transform the test messages into feature vectors using the same vocabulary as the training data
message_test_features = feature_extraction.transform(message_test)

# Convert category_train and category_test to integers if they are not already
category_train = category_train.astype(int)
category_test = category_test.astype(int)
```

```python
print(message_train)
```

```python
print(message_train_features)
```

```python
model=LogisticRegression()
```

```python
model.fit(message_train_features, category_train)
```

```python
prediction_on_training_data=model.predict(message_train_features)
accuracy_on_training_data=accuracy_score(category_train,prediction_on_training_data)
print("accuracy on training data: " , accuracy_on_training_data)
```

```python
prediction_on_testing_data=model.predict(message_test_features)
accuracy_on_testing_data=accuracy_score(category_test,prediction_on_testing_data)
print("accuracy on testing data: " , accuracy_on_testing_data)
```

```python
# using random forest algorithm we get more accuarcy
from sklearn.ensemble import RandomForestClassifier

# Initialize the Random Forest classifier
rf_classifier = RandomForestClassifier(n_estimators=100, random_state=42)

# Train the Random Forest classifier
rf_classifier.fit(message_train_features, category_train)

# Predict on the test data
category_test_predicted = rf_classifier.predict(message_test_features)
```
