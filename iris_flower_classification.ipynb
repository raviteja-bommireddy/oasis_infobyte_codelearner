```python
data=df.values
X=data[:,0:4]
Y=data[:,4]
print(X)
print(Y)
```

```python
X = pd.DataFrame(X)
X.corr()
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test= train_test_split(X,Y,test_size=0.3)
print(X_test)
print(Y_test)
```

```python
from sklearn.svm import SVC
model_svc=SVC()
model_svc.fit(X_train,Y_train)
```

```python
prediction_of_model_1=model_svc.predict(X_test)
from sklearn.metrics import accuracy_score
print(accuracy_score(Y_test,prediction_of_model_1))
```

```python
from sklearn.linear_model import LogisticRegression
model_LR=LogisticRegression()
model_LR.fit(X_train,Y_train)
```

```python
prediction_of_model_2=model_LR.predict(X_test)
from sklearn.metrics import accuracy_score
print(accuracy_score(Y_test,prediction_of_model_2))
```

```python
from sklearn.tree import DecisionTreeClassifier
model_DTC=DecisionTreeClassifier()
model_DTC.fit(X_train,Y_train)
```

```python
prediction_of_model_3=model_DTC.predict(X_test)
from sklearn.metrics import accuracy_score
print(accuracy_score(Y_test,prediction_of_model_3))
```

```python
from sklearn.ensemble import RandomForestClassifier
model_RFC=RandomForestClassifier()
model_RFC.fit(X_train,Y_train)
```

```python
prediction_of_model_4=model_RFC.predict(X_test)
from sklearn.metrics import accuracy_score
print(accuracy_score(Y_test,prediction_of_model_4))
```

```python
from sklearn.metrics import classification_report
print(classification_report(Y_test,prediction_of_model_1))
```
